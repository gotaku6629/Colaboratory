**データセットからの画像の視覚化**

> 今回は, 複数の画像を1枚の画像の中で表示させる！
cf) http://ailaby.com/matplotlib_fig/



```
import numpy as np
import matplotlib.pyplot as plt
import random

def show_some_examples(names, data, labels):
  plt.figure()   #大きなfigureをまず1つ確保
  rows, cols = 4, 4
  random_idxs = random.sample(range(len(data)), rows * cols) #(data, num):dataの中からnum個を取り出す.

  for i in range(rows * cols):
    plt.subplot(rows, cols, i+1)  #figure内を分割:(行数, 列数, 何番目に配置するか)
    j = random_idxs[i]
    plt.title(names[labels[j]])   #subplotのタイトル
    img = np.reshape(data[j, :], (24, 24))
    plt.imshow(img, cmap='Greys_r')  #行列を貼り付けて,表示！
    plt.axis('off')
  
  plt.tight_layout() #重ならないように調整
  plt.savefig('cifar_examples.png')  #保存
```

```
show_some_examples(names, data, labels)
```
